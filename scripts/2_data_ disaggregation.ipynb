# Disaggregating the data

Team Sheffield felt a more natural format for the data was to *disaggregate* it. In other words, to unpack each dyad into singular ref-player interaction. In other words, each time a player and a referee met they would contribute one row to the dataset (and thus that interaction has a maximum of one red card). Based on https://osf.io/w7tds/

```python
import pandas as pd # for dealing with csv import
import os # for joining paths and filenames sensibly
```

```python
# set wkdir
file_path = '/Users/sebastian/Documents/Uni/Sheffield (MSc)/2. Semester/Research Project/msc_dissertation_project'
os.chdir(file_path)
```

```python
# nb new datase
filename = os.path.join('data', '1_crowdstorming_data.csv') 
df = pd.read_csv(filename)
```

```python
# add new vars
df['skintone'] = (df['rater1'] + df['rater2']) / 2
df['allreds'] = df['yellowReds'] + df['redCards']
df['allredsStrict'] = df['redCards']
df['refCount'] = 0
```

```python
# inspect data
pd.set_option('display.max_columns', None)
pd.DataFrame.head(df)
```

```python
# add a column which tracks how many games each ref is involved in
refs = pd.unique(df['refNum'].values.ravel()) # list all unique ref IDs
```

```python
# for each ref, count their dyads
for r in refs:
    df['refCount'][df['refNum'] == r] = len(df[df['refNum'] == r])

# for r in refs:
#     if len(df[df['refNum'] == refs[r]]) > 1:
#        print len(df[df['refNum'] == refs[r]])
```

```python
colnames = list(df.columns)
```

```python
j = 0
out = [0 for _ in range(sum(df['games']))]

for _, row in df.iterrows():
        n = row['games']
        c = row['allreds']
        d = row['allredsStrict']
        
        # row['games'] = 1        
        
        for _ in range(n):
                row['allreds'] = 1 if (c-_) > 0 else 0
                row['allredsStrict'] = 1 if (d-_) > 0 else 0
                # convert from pandas Series to prevent overwriting previous values of out[j]
                rowlist = list(row)
                out[j] = rowlist
                j += 1
                if j% 10 == 0:    
                    print("Number " + str(j) + " of " + str(df.shape[0]))
```

```python
# inspect data
pd.set_option('display.max_columns', None)
pd.DataFrame(out, columns = colnames)
```

```python
pd.DataFrame(out, columns = colnames).to_csv(file_path + '/data/2_disaggregated_data.csv', index = False)
```
